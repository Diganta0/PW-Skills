Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Answer:

##Underfitting
- The model is too simple to capture the underlying patterns in the data. It performs poorly on both the training data and the test data.

- Bias/Variance: It has high bias (it makes strong, simple assumptions) and low variance.


##Overfitting
- The model is too complex and learns the training data too well, including the noise. It performs perfectly on the training data but fails badly on new, unseen test data.

- Bias/Variance: It has low bias and high variance (it's highly sensitive to small changes in the training data).


##Detection

- Overfitting: High accuracy on training data (e.g., 99%) but low accuracy on test data (e.g., 70%).

- Underfitting: Low accuracy on both training data (e.g., 60%) and test data (e.g., 62%).


##Prevention
###Preventing Underfitting (High Bias):

1. Use a more complex model: Switch from linear to a polynomial model, or use a more powerful algorithm (e.g., a Random Forest instead of a simple Decision Tree).

2. Add more features: The model may be simple because it doesn't have enough information.

3. Train longer: Sometimes the model just hasn't had enough time to learn.

###Preventing Overfitting (High Variance):

1. Get more data: The best way. More data helps the model learn the true pattern and ignore the noise.

2. Cross-Validation: A technique (like k-fold cross-validation) that splits the data into many train/test sets. It gives a more reliable estimate of how the model will perform on new data.

3. Regularization: A technique that penalizes model complexity.

- L1 (Lasso): Adds a penalty equal to the absolute value of the weights. Can shrink some weights to zero, effectively doing feature selection.

- L2 (Ridge): Adds a penalty equal to the square of the weights. Keeps all weights small but doesn't make them zero.

4. Feature Selection: Remove irrelevant features that are just adding noise.

5. Model Simplification: (e.g., "Pruning" a decision tree, using "Dropout" in a neural network).

Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.

Answer:

Here are three common methods to handle missing values (NaNs) in a dataset:

##Deletion (Listwise Deletion)
1. Explanation: This method involves simply deleting any row that contains one or more missing values.

2. Example: If you have a dataset of 100 users, but 5 users are missing their 'Age', you delete those 5 rows, leaving you with a dataset of 95 users.

3. Code: df_cleaned = df.dropna()

4. When to use: Good for large datasets where losing a few rows won't cause a problem or introduce bias.

5. Problem: You lose data, which can reduce your model's power. If the missing data isn't random (e.g., only high-income users are missing 'Age'), this can bias your model.

##Mean/Median/Mode Imputation
1. Explanation: This method involves replacing the missing value with a statistical value calculated from the rest of the column.

- Mean: Use the average value. Good for normally distributed data.

- Median: Use the middle value. Better for data with outliers (skewed data), like 'Salary'.

- Mode: Use the most frequent value. Used for categorical (non-numeric) columns, like 'Color'.

2. Example: If the 'Age' column is missing a value and the median age of all other users is 32, you fill the missing spot with 32.

3. Code: median_age = df['Age'].median() df['Age'] = df['Age'].fillna(median_age)

4. When to use: Quick and easy. Prevents data loss.

5. Problem: It reduces the variance of the data and can distort the relationships between variables.

##Predictive Modeling (e.g., KNN Imputer)
1. Explanation: This is a more advanced method where you use other features in the dataset to predict what the missing value should be.

- KNN Imputer: Finds the 'k' nearest neighbors (users who are most similar based on other features like 'Salary' and 'Education') and uses their 'Age' values to guess the missing 'Age'.

- Regression: You could build a linear regression model Age = m(Salary) + c to predict the missing age.

2. Example: A user is missing 'Age' but has 'Salary' = $50,000 and 'Education' = 'Masters'. The KNN algorithm finds 3 other users with similar salaries and education, whose ages are 28, 29, and 30. It would then fill the missing value with the average, 29.

3. When to use: When data is valuable and you want the most accurate guess for the missing value.

4. Problem: It is much more complex and computationally expensive than simple imputation.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).

Answers:

 An imbalanced dataset is a classification problem where one class (the majority class) has a very large number of samples, while another class (the minority class) has very few.

- Example: A credit card fraud dataset.

1. 99.9% of transactions are 'Not Fraud' (majority class).

2. 0.1% of transactions are 'Fraud' (minority class).

This is a problem because a "dumb" model can achieve 99.9% accuracy by always predicting 'Not Fraud', but it completely fails at its real job: catching the fraud.

Here are two techniques to handle it:

##SMOTE (Oversampling)
Theory (Theoretical): SMOTE stands for Synthetic Minority Over-sampling TEchnique. Instead of just duplicating existing minority samples (which can lead to overfitting), SMOTE creates new, synthetic samples.

1. It takes a sample from the minority class (e.g., a 'Fraud' transaction).

2. It finds its 'k' nearest neighbors (other 'Fraud' transactions that are similar).

3. It randomly picks one of those neighbors.

4. It creates a new synthetic sample at a random point along the line between the original sample and its chosen neighbor.

##Random Undersampling
- Theory (Theoretical): This is the opposite of oversampling. Instead of making the minority class bigger, it makes the majority class smaller. It does this by randomly deleting samples from the majority class until the dataset is balanced (e.g., 50% 'Fraud' and 50% 'Not Fraud').
- When to use/Problem: This is very fast, but its main drawback is that you lose a lot of data from the majority class, which might contain important information. It's best used when you have a very large dataset.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Answers:

1. Distance-Based Algorithms (e.g., KNN, SVM, K-Means): These algorithms work by calculating the "distance" between data points. If one feature (e.g., Salary ranging from 50,000 to 500,000) is on a much larger scale than another (e.g., Age ranging from 20 to 60), the Salary feature will completely dominate the distance calculation. The Age feature will be ignored. Scaling ensures all features contribute fairly.

2. Gradient Descent (e.g., Linear Regression, Neural Networks): This is the algorithm used to train (optimize) many models. It works by finding the "bottom" of an error "hill." If features are on different scales, this hill becomes long and narrow (an oval shape). Gradient descent will take a long, slow, zigzag path to find the bottom. Scaling makes the hill more like a circle, allowing the algorithm to converge to the best solution much faster and more reliably.


| Feature | Min-Max Scaling (Normalization) | Standardization |
|-----------|-----------|-----------|
| Formula | $X_{\text{scaled}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}$ | $X_{\text{scaled}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}$$X_{\text{scaled}} = \frac{X - \mu}{\sigma}$ (where $\mu$ = mean, $\sigma$ = std. dev) |
| Output Range | Scales data to a fixed range, usually [0, 1]. | Centers the data around a mean of 0 with a standard deviation of 1. |
| Effect of Outliers | Very sensitive to outliers. A single huge outlier can squash all the other data into a tiny range (e.g., 0 to 0.1). | Much more robust to outliers. The mean and standard deviation are less affected by a few outliers. |
| When to Use | Good for algorithms that require data in a specific range (e.g., [0, 1] for neural networks). Also used in image processing. | The default "go-to" scaling method. Used for most algorithms (SVM, Logistic Regression) that assume data is centered at zero. |

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

Answers:

Both methods are used to convert categorical string data (like "Red", "Green", "Blue") into numbers so machine learning models can understand them.

##Label Encoding
1. What it is: It assigns a unique integer to each category.

2. Example:

  - 'Low' -> 0
  - 'Medium' -> 1
  - 'High' -> 2

3. Problem: The model might learn a false ordinal relationship. It will think High (2) > Medium (1), which is true in this case. But if you encode [India, USA, Japan] as [0, 1, 2], the model might learn that Japan > USA, which is meaningless.

4. When to use: Only for Ordinal Data. This is data that has a clear, built-in order (e.g., 'Low', 'Medium', 'High'; 'Bad', 'Good', 'Excellent'; 'PhD', 'Masters', 'Bachelors').

## One-Hot Encoding
1. What it is: It creates new binary (0 or 1) columns for each unique category.

2. Benefit: It breaks the false ordinal relationship. The model now just sees which category is "on" (1) or "off" (0), with no category being "larger" than another.

3. When to use: For Nominal Data. This is data with no intrinsic order (e.g., 'Color', 'Country', 'City'). This is the safest and most common encoding method.

4. Problem: If your feature has many unique categories (e.g., 50 different 'States'), it will create 50 new columns, which can make your dataset very large (this is called the "curse of dimensionality").
