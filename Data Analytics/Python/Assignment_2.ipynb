# Data Types and Structures





1. What are data structures, and why are they important?
  - Data structures are ways to store and organize data in a computer so it can be used efficiently. They are important because they help in solving problems and managing large amounts of data.

2. Explain the difference between mutable and immutable data types with examples.
  - Mutable: You can change them after they are created, like a list.
  - Immutable: You cannot change them after they are created, like a string.

3. What are the main differences between lists and tuples in Python?
  - Lists can be changed (mutable), while tuples cannot be changed (immutable). Lists use [] and tuples use ().

4. Describe how dictionaries store data.
  - Dictionaries store data in key-value pairs. Each key is unique and is used to find its value.

5. Why might you use a set instead of a list in Python?
  - You would use a set to automatically handle duplicate values, as sets only store unique items.

6. What is a string in Python, and how is it different from a list?
  - A string is a sequence of characters. It is different from a list because a string is immutable, while a list is mutable.

7. How do tuples ensure data integrity in Python?
  - Tuples are immutable, so their elements cannot be changed once created. This prevents accidental changes to the data.

8. What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table is a data structure used for fast data lookup. Dictionaries in Python are built using hash tables.

9. Can lists contain different data types in Python?
  - Yes, lists can contain different data types.

10. Explain why strings are immutable in Python.
  - Strings are immutable to make them reliable and safe. Their value cannot be changed after creation.

11. What advantages do dictionaries offer over lists for certain tasks?
  - Dictionaries offer faster data lookup because you can access values directly by their unique keys.

12. Describe a scenario where using a tuple would be preferable over a list.
  - Using a tuple is better when you have a collection of items that should not change, like the coordinates of a point on a map.

13. How do sets handle duplicate values in Python?
  - Sets automatically remove duplicate values.

14. How does the "in" keyword work differently for lists and dictionaries?
  - For lists, it checks if a value exists within the list's values. For dictionaries, it checks if a value exists as a key.

15. Can you modify the elements of a tuple? Explain why or why not.
  - No, you cannot modify the elements of a tuple because they are immutable.

16. What is a nested dictionary, and give an example of its use case?
  - A nested dictionary is a dictionary inside another dictionary. A use case is storing information about students, where each student's data is a separate dictionary.

17. Describe the time complexity of accessing elements in a dictionary.
  - The time complexity of accessing elements in a dictionary is very fast, close to constant time, or O(1).

18. In what situations are lists preferred over dictionaries?
  - Lists are preferred when you have an ordered collection of items or need to access items by their position (index).

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
  -Dictionaries are considered unordered because the order of items is not guaranteed. This doesn't affect data retrieval because you access data using keys, not by their position.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
  - Lists: You retrieve items by their index (position).
  - Dictionaries: You retrieve items by their key.
