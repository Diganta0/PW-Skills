1. What is the difference between a function and a method in Python?

2. Explain the concept of function arguments and parameters in Python.
  - A parameter is a variable in the function definition that holds a place for a value. An argument is the actual value you pass to the function when you call it.
  - Example:<br>
    def greet(name): # 'name' is the parameter<br>
    print(f"Hello, {name}!")<br>
    greet("Diganta") # "Diganta" is the argument<br>


3. What are the different ways to define and call a function in Python?

  - Example:<br>
    def add(x, y): # Defining the function<br>

4. What is the purpose of the return statement in a Python function?

  - Example:<br>
    def get_pi():<br>

5. What are iterators in Python and how do they differ from iterables?
  - An iterable is an object you can loop over (like a list). An iterator is an object that keeps track of the current position while looping and gives you the next item.
  - Example:<br>
    my_list = [1, 2, 3] # This is an iterable<br>
    my_iter = iter(my_list) # This is an iterator<br>
    print(next(my_iter)) # Output: 1<br>
    print(next(my_iter)) # Output: 2<br>


6. Explain the concept of generators in Python and how they are defined.
  - A generator is a special type of function that uses the yield keyword to create an iterator. It produces values one at a time, saving memory.
  - Example:<br>
    def my_generator():<br>
      yield 1<br>
      yield 2<br>
    gen = my_generator()<br>
    print(next(gen)) # Output: 1<br>


7. What are the advantages of using generators over regular functions?
  - Generators are memory-efficient because they produce values on demand instead of creating a full list in memory. They are also useful for working with large data sets.
  - Example:<br>
    A generator for a large range of numbers (memory efficient)<br>
    numbers = (i for i in range(1000000))<br>
    A list for the same range (uses more memory)<br>
    numbers_list = [i for i in range(1000000)]<br>


8. What is a lambda function in Python and when is it typically used?
  - A lambda function is a small, anonymous function with no name. It is used for simple, one-line operations and is often passed as an argument to other functions.

  - Example:<br>
    A lambda function to add 10 to a number<br>
    add_ten = lambda x: x + 10<br>
    print(add_ten(5)) # Output: 15<br>


9. Explain the purpose and usage of the map() function in Python.
  - The map() function applies a specific function to every item in an iterable. It's used to transform data quickly.
  - Example:<br>
    numbers = [1, 2, 3, 4]<br>
    squared = list(map(lambda x: x**2, numbers))<br>
    print(squared) # Output: [1, 4, 9, 16]<br>


10. What is the difference between map(), reduce(), and filter() functions in Python?
  - map() applies a function to every item.
  - filter() keeps only the items that meet a certain condition.
  - reduce() combines all items into a single value.
  - Example:<br>
    from functools import reduce<br>
    numbers = [1, 2, 3, 4, 5]<br><br>
    
  ##### map: double each number<br>
    doubled = list(map(lambda x: x * 2, numbers))<br>
    print(f"map result: {doubled}") #output [2, 4, 6, 8, 10]<br><br>

  ##### filter: get only even numbers<br>
    evens = list(filter(lambda x: x % 2 == 0, numbers))<br>
    print(f"filter result: {evens}") #output [2, 4]<br><br>

  ##### reduce: sum all numbers<br>
    total = reduce(lambda x, y: x + y, numbers)<br>
    print(f"reduce result: {total}") #output 15<br>

```python
#PRACTICAL





#1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_evens(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

my_list = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(my_list))


#2 Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(s):
    return s[::-1]

my_string = "hello"
print(reverse_string(my_string))


#3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def get_squares(numbers):
    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares

my_list = [1, 2, 3, 4]
print(get_squares(my_list))


#4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(is_prime(17))
print(is_prime(18))


#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration

        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num, end=" ")


#6 Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
    power = 1
    for _ in range(n + 1):
        yield power
        power *= 2

for p in power_of_two(5):
    print(p, end=" ")


#7 Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file):
    for line in file:
        yield line.strip()

mock_file = ["Line 1\n", "Line 2\n", "Line 3\n"]
for line in read_file_lines(mock_file):
    print(line)


#8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
students = [('John', 'A', 15), ('Jane', 'B', 12), ('Dave', 'B', 10)]

sorted_students = sorted(students, key=lambda student: student[1])
print(sorted_students)


#9 Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temps = [0, 10, 20, 30, 40]

fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print(fahrenheit_temps)


#10 Create a Python program that uses filter() to remove all the vowels from a given string.
def is_consonant(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

my_string = "Hello World"
result = "".join(filter(is_consonant, my_string))
print(result)


#11 Imagine an accounting routine used in a book shop. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €. Write a Python program using lambda and map.
orders = [
    ("34587", "Learning Python, Mark Lutz", 4, 40.95),
    ("98762", "Programming Python, Mark Lutz", 5, 56.80),
    ("77226", "Head First Python, Paul Barry", 3, 32.95),
    ("88112", "Einführung in Python3, Bernd Klein", 3, 24.99)
]

result_list = list(map(lambda item: (item[0], item[2] * item[3] + 10) if item[2] * item[3] < 100 else (item[0], item[2] * item[3]), orders))
print(result_list)
```
