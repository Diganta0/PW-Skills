1. What is a random variable in probability theory?
  - A random variable is a variable whose value is a numerical outcome of a random event. It connects the outcomes of an experiment (which can be words or symbols) to a number.
  - Example: If you flip a coin, the outcomes are "Heads" or "Tails". You can define a random variable X where X = 1 if it's Heads and X = 0 if it's Tails. The variable X represents the outcome with a number.

2. What are the types of random variables?
  - There are two main types of random variables
  - Discrete Random Variable: Can only take specific, countable values. You can't have a value in between.
  - Example: The number rolled on a die (can only be 1, 2, 3, 4, 5, or 6) or the number of students in a class.

  - Continuous Random Variable: Can take any value within a given range.
  - Example: A person's height (can be 175cm, 175.1cm, 175.11cm, etc.) or the temperature in a room.

3. Explain the difference between discrete and continuous distributions.
  - Discrete Distribution: Describes the probability of countable outcomes. Its graph is a bar chart where each bar represents the probability of a specific value.
  - Example: The probability of getting 1, 2, 3, 4, 5, or 6 when rolling a die.

  - Continuous Distribution: Describes the probability of outcomes over a continuous range. Its graph is a smooth curve. You find the probability for a range of values (e.g., probability of height between 170cm and 180cm), not a single point.
  - Example: The distribution of people's weights.

4. What is a binomial distribution, and how is it used in probability?
  - A binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes.
  - It applies when these three conditions are met:
  - There is a fixed number of trials (e.g., flipping a coin 10 times).
  - Each trial has only two outcomes (e.g., heads or tails).
  - The trials are independent, and the probability of success is the same for each trial.
  - How it's used: It helps calculate the probability of getting a specific number of successes.
  - Example: You can use it to find the probability of a new drug curing exactly 8 out of 10 patients.

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that if you take many large enough random samples from any population (even a non-normal one) and calculate the mean of each sample, the distribution of those sample means will form a normal distribution (a bell curve).

  - The CLT is the foundation of inferential statistics. It lets us use statistical methods that require a normal distribution (like confidence intervals and hypothesis tests), even if we don't know the original population's distribution. As long as our sample size is large enough (usually n > 30), we can assume the sample means will be normally distributed.

7. What is the significance of confidence intervals in statistical analysis?
  - A confidence interval is a range of values, calculated from sample data, that likely contains the true value of a population parameter (like the population mean).

  - Its significance is that it quantifies uncertainty. Instead of giving a single-number estimate (a point estimate), which is almost certainly wrong, it provides a range.

  - Example: Instead of saying "Average sales are â‚¹250," you say "We are 95% confident that the true average sales are between â‚¹240 and â‚¹260."
  - This is crucial for decision-making. A narrow interval implies a precise estimate, while a wide interval tells you that you need more data to be certain.

8. What is the concept of expected value in a probability distribution?
  - The expected value is the long-term average outcome you would expect if you repeated an experiment many times. It's a weighted average of all possible outcomes, where each outcome is weighted by its probability.

  - Example: For a fair six-sided die, the possible outcomes are 1, 2, 3, 4, 5, 6. The expected value is 3.5. You can never roll a 3.5, but if you rolled the die thousands of times and averaged the results, your average would be very close to 3.5.

```python
# Q9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation, and draw a histogram.
# !pip install numpy scipy matplotlib -q
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the distribution
mean = 50
std_dev = 5
num_samples = 1000

# Generate random numbers from a normal distribution
random_data = np.random.normal(mean, std_dev, num_samples)

# Compute the mean and standard deviation of the generated sample
sample_mean = np.mean(random_data)
sample_std_dev = np.std(random_data)

# Print the results
print(f"Generated {num_samples} numbers from a normal distribution.")
print(f"Original Mean: {mean}, Original Std Dev: {std_dev}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Std Dev: {sample_std_dev:.2f}\n")

# Visualize the distribution with a histogram
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram of Generated Normal Distribution Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

10. You are a data analyst. Explain how to apply the CLT to estimate the average sales with a 95% confidence interval using the provided data. Write the Python code.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

  - How the Central Limit Theorem (CLT) Applies: We only have a small sample (20 days) of sales data. The CLT lets us assume that the distribution of sample means would be approximately normal, even if we don't know the distribution of the daily sales for the full two years. This critical assumption allows us to calculate a valid confidence interval.

  - 95% Confidence Interval: Using this assumption, we will calculate a 95% confidence interval. This will give us a range of values where we can be 95% confident that the true average daily sales for the entire two-year period lies.

```python
import numpy as np
from scipy import stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Calculate sample statistics
sample_mean = np.mean(daily_sales)
sample_std = np.std(daily_sales, ddof=1) # ddof=1 for sample standard deviation
n = len(daily_sales)

# Calculate the Standard Error of the Mean (SEM)
sem = sample_std / np.sqrt(n)

# Calculate the 95% confidence interval using the t-distribution
# We use the t-distribution because the sample size is small (n < 30)
confidence_level = 0.95
degrees_freedom = n - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sem)

# Print the results
print(f"Sample Mean Sales: â‚¹{sample_mean:.2f}")
print(f"95% Confidence Interval for True Average Sales: (â‚¹{confidence_interval[0]:.2f}, â‚¹{confidence_interval[1]:.2f})")
```
